# Effects of Artificial Intelligence on Human Jobs

## Introduction 

#### What is Artificial Intelligence 

##### Artificial intelligence is when computers or machines are designed to perform tasks that usually require human intelligence. This could include things like understanding language, recognizing pictures, playing games, or even driving cars. Unlike regular software that only follows strict instructions, AI can learn from experience and improve over time.

#### Why to study impacts of Artificial Intelligence on Human Jobs

##### 1. Economic Implications
###### Job Displacement vs. Job Creation: AI automation can replace routine and repetitive tasks, potentially displacing millions of workers. Simultaneously, it generates new roles in AI development, maintenance, and oversight. Quantifying these shifts is essential for policymakers and businesses to mitigate unemployment and skill mismatches.
###### Productivity and Economic Growth: AI integration enhances efficiency and productivity. Understanding which sectors are most affected helps allocate resources to maximize economic benefits without adversely impacting labor markets.

##### 2. Workforce Transformation
###### Reskilling and Upskilling Needs: Identifying labor displacement patterns enables governments and organizations to design effective reskilling programs, upskill workers, and prepare them for AI-augmented roles.
###### Changing Skill Demands: AI may reduce demand for manual or repetitive cognitive tasks but increase demand for creative, strategic, and emotional intelligence skills. Studying trends allows proactive workforce adaptation.
##### 3. Social and Ethical Considerations
###### Income Inequality: Without mitigation, AI-induced job displacement can exacerbate wealth and social inequalities. Research provides data to inform equitable policies such as universal basic income, targeted social safety nets, and fair wage practices.
###### Psychological Impact: Unemployment or job uncertainty due to automation can affect mental health and social cohesion. Studying AI’s impact helps in devising supportive infrastructures and workplace transition strategies.

##### 4. Policy and Regulatory Guidance
###### Legislation and Labor Protection: Understanding AI’s effect on employment allows policymakers to design laws that protect workers’ rights and ensure ethical deployment of AI in workplaces.
###### Strategic Planning for Future Work: Governments can strategically plan labor markets for long-term sustainability, balancing AI adoption with human job preservation.
##### 5. Technological Design and Societal Integration
###### Human-in-the-Loop Systems: Research facilitates designing AI systems where humans and machines complement each other rather than compete, enhancing collaboration and safety.
###### Inclusive Development: Studying labor impacts ensures AI technologies are deployed in ways that benefit society broadly rather than disproportionately displacing specific demographics.

#### Some charts for more simple understanding : 

```python
import matplotlib.pyplot as plt
```

```python
c1 = df.sort_values(
    by="AI_Exposure_Index",
    ascending = False
).head(10)

plt.figure(figsize=(20,5))
plt.bar(
    c1["Job_Title"],-
    c1["AI_Exposure_Index"]
)
plt.title("TOP RISK PROFILES DUE TO AI")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.hist(df["Average_Salary"],bins=25)
plt.title("Salary Distribution")
plt.xlabel("SALARY")
plt.ylabel("COUNT")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.scatter(df["Average_Salary"],df["AI_Exposure_Index"])
plt.title("AVERAGE SALARY VS AI EXPOSURE")
plt.xlabel("SALARY")
plt.ylabel("AI EXPOSURE")
plt.show()
```

```python
c3 = df.sort_values(
    by = "Automation_Probability_2030",
    ascending = False
).head(5)

plt.figure(figsize = (5,8))

plt.pie(
    c3["Automation_Probability_2030"],
    labels = c3["Job_Title"],
    autopct = "%1.1f%%"
       )

plt.show()
```
